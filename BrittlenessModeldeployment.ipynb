## WWell_1920837

```python
WWell_1920837prepared=pd.read_csv(r'C:\Users\eobasi\OneDrive - University of Wyoming\Desktop\PowderRiverBasin Project work\Geologicalsweetspotmap\WWell_1920837prepared.csv')
```

```python
WWell_1920837prepared
```
